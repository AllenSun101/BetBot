```python
import pandas as pd
from matplotlib import pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
team_stats = pd.read_csv("../team_stats.csv")
player_stats = pd.read_csv("../player_stats.csv")
```

```python
team_stats.head()
```

```python
plt.hist(team_stats["passing_yards"])
plt.title("Passing Yards")
```

```python
plt.hist(team_stats["rushing_yards"])
plt.title("Rushing Yards")
```

```python
plt.scatter(team_stats["rushing_yards"], team_stats["passing_yards"])
plt.title("Pass Yards vs. Rush Yards")
```

```python
plt.scatter(team_stats["completions"] / team_stats["attempts"], team_stats["passing_yards"])
plt.title("Pass Yards vs. Pass Completion Rate")
```

```python
plt.scatter(team_stats["completions"] / team_stats["attempts"], team_stats["passing_epa"])
plt.title("Passing EPA vs. Pass Completion Rate")
```
